```python
import pandas as pd
import matplotlib.pyplot as plt
import talib
import seaborn as sns

# Load the CSV file for AMZN_historical_data

df_AMZN = pd.read_csv('AMZN_historical_data.csv')  # Adjust the path

# Convert 'Date' to datetime format
df_AMZN['Date'] = pd.to_datetime(df_AMZN['Date'])

# Sort data by date to ensure time series correctness
df_AMZN.sort_values('Date', inplace=True)

# Set 'Date' as index to make it time-series friendly
df_AMZN.set_index('Date', inplace=True)

# Display the first few rows to understand the structure
df_AMZN.head()
```

```python
# Load the CSV file for GOOG_historical_data

df_GOOG = pd.read_csv('GOOG_historical_data.csv')  # Adjust the path

# Convert 'Date' to datetime format
df_GOOG['Date'] = pd.to_datetime(df_GOOG['Date'])

# Sort data by date to ensure time series correctness
df_GOOG.sort_values('Date', inplace=True)

# Set 'Date' as index to make it time-series friendly
df_GOOG.set_index('Date', inplace=True)

# Display the first few rows to understand the structure
df_GOOG.head()
```

```python
# Load the CSV file for META_historical_data

df_META = pd.read_csv('META_historical_data.csv')  # Adjust the path

# Convert 'Date' to datetime format
df_META['Date'] = pd.to_datetime(df_META['Date'])

# Sort data by date to ensure time series correctness
df_META.sort_values('Date', inplace=True)

# Set 'Date' as index to make it time-series friendly
df_META.set_index('Date', inplace=True)

# Display the first few rows to understand the structure
df_META.head()
```

```python
# Calculate 20-day Simple Moving Average (SMA) for df_AMZN
df_AMZN['SMA_20'] = talib.SMA(df_AMZN['Close'], timeperiod=20)

# Calculate Relative Strength Index (RSI) with 14-day window
df_AMZN['RSI_14'] = talib.RSI(df_AMZN['Close'], timeperiod=14)

# Calculate MACD (Moving Average Convergence Divergence)
df_AMZN['MACD'], df_AMZN['MACD_Signal'], df_AMZN['MACD_Hist'] = talib.MACD(df_AMZN['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
```

```python
# Calculate 20-day Simple Moving Average (SMA) for df_GOOG
df_GOOG['SMA_20'] = talib.SMA(df_GOOG['Close'], timeperiod=20)

# Calculate Relative Strength Index (RSI) with 14-day window
df_GOOG['RSI_14'] = talib.RSI(df_GOOG['Close'], timeperiod=14)

# Calculate MACD (Moving Average Convergence Divergence)
df_GOOG['MACD'], df_GOOG['MACD_Signal'], df_GOOG['MACD_Hist'] = talib.MACD(df_GOOG['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
```

```python
# Calculate 20-day Simple Moving Average (SMA) for df_META
df_META['SMA_20'] = talib.SMA(df_META['Close'], timeperiod=20)

# Calculate Relative Strength Index (RSI) with 14-day window
df_META['RSI_14'] = talib.RSI(df_META['Close'], timeperiod=14)

# Calculate MACD (Moving Average Convergence Divergence)
df_META['MACD'], df_META['MACD_Signal'], df_META['MACD_Hist'] = talib.MACD(df_META['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
```

```python
# Calculate Daily Returns df_AMZN
daily_returns = df_AMZN['Close'].pct_change().dropna()

# Calculate Mean Return
mean_return = daily_returns.mean()

# Calculate Volatility (Standard Deviation)
volatility = daily_returns.std()

# Calculate Sharpe Ratio (assuming risk-free rate ≈ 0 for simplicity)
sharpe_ratio = mean_return / volatility

# Display the results
print(f"Mean Daily Return: {mean_return:.4f}")
print(f"Volatility (Standard Deviation): {volatility:.4f}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
```
